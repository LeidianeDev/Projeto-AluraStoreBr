### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = sum(loja['Preço'])
faturamento_loja2 = sum(loja2['Preço'])
faturamento_loja3 = sum(loja3['Preço'])
faturamento_loja4 = sum(loja4['Preço'])
```

```python
print(f'Faturamento da Loja 1: R${faturamento_loja1:.2f}')
print(f'Faturamento da Loja 2: R${faturamento_loja2:.2f}')
print(f'Faturamento da Loja 3: R${faturamento_loja3:.2f}')
print(f'Faturamento da Loja 4: R${faturamento_loja4:.2f}')
```

Gráfico

```python
lojas = ['loja', 'loja2', 'loja3', 'loja4']
faturamentos = [1534509.12, 1488459.06, 1464025.03, 1384497.58]

plt.bar(lojas, faturamentos) #tipo de gráfico
plt.title('faturamento')#título

for i, lojas in enumerate(faturamentos):
    plt.text(i, lojas + 1, f'{faturamentos[i]:.1f}R$', ha='center')

plt.xlabel('lojas') #descrição eixo x comprimento
plt.ylabel('faturamento(R$ Milhão)') #descrição eixo y altura
plt.show()
```

## Descrição do Projeto – Análise de Lojas da Alura Store

Este projeto tem como objetivo realizar uma análise exploratória dos dados de quatro lojas virtuais fictícias pertencentes à rede Alura Store. A intenção principal é avaliar o desempenho individual de cada loja com base em indicadores-chave, de forma a embasar uma decisão estratégica: qual loja deve ser vendida para viabilizar um novo investimento do Sr. João.

Para ajudar o Sr. João a decidir qual loja da Alura Store vender, foi realizada uma análise detalhada considerando os seguintes critérios:


*   **Faturamento:** Verificamos qual loja gera mais receita e qual tem baixo retorno financeiro.
*   **Produtos:** Analisamos a variedade, desempenho e ticket médio dos produtos vendidos.


*  **Avaliações de clientes:** Consideramos a média das notas atribuídas pelos consumidores para entender a percepção de qualidade.
*   **Frete médio:** Avaliamos o custo médio do frete, fator que pode influenciar na satisfação do cliente e nas vendas.

##🧠 Análise

**Loja 1:** Melhor desempenho em quase todos os indicadores. Maior faturamento, melhor ticket médio e avaliação mais alta.

**Loja 2:** Levemente abaixo da Loja 1, mas ainda com bons números e desempenho equilibrado.

**Loja 3:** Desempenho intermediário. Avaliação um pouco menor, ticket médio mais baixo que as anteriores.

**Loja 4:** Apresenta os piores resultados em todos os indicadores:

Menor faturamento total.

Menor avaliação média.

Menor ticket médio.

✅ Recomendação:

###O Sr. João deve considerar vender a Loja 4.

  ## Justificativa:

**Menor faturamento:** R$ 1.384.497,58 – significativamente abaixo das demais.

**Pior avaliação média:** 3,98 – indica menor satisfação dos clientes.

**Menor ticket médio:** R$ 587,15 – indica menor valor médio por venda.

**Custo de frete mais baixo:** embora seja positivo, não compensa o fraco desempenho nos demais indicadores.

Essa loja representa o menor retorno financeiro e menor percepção de valor pelos clientes. Portanto, é a opção mais estratégica para ser descontinuada, permitindo que o Sr. João foque em operações mais rentáveis e com maior potencial de crescimento.

A partir dessa análise, identificamos a loja 4 com o pior desempenho geral — considerando baixa eficiência, avaliações negativas e retorno financeiro insatisfatório
